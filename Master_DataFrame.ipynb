```python
movie_joined = pd.read_sql("""
SELECT
    movie_id,
    primary_title, 
    start_year, 
    runtime_minutes, 
    genres,
    averagerating, 
    numvotes
FROM movie_basics
JOIN movie_ratings
    USING(movie_id)
""", conn)
```

```python
movie_joined.info()
```

```python
profession_joined = pd.read_sql("""
SELECT
    movie_id,
    category, 
    primary_name
FROM principals
JOIN persons
    USING(person_id)
""", conn)


profession_joined.info()
```

```python
profession_joined = profession_joined[(profession_joined['category'] == 'actor') | (profession_joined['category'] == 'director')]
```

```python
movie_profession = movie_joined.merge(profession_joined)
```
