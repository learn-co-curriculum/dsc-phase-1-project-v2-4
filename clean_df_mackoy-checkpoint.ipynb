```python
import pandas as pd
import sqlite3
import numpy as np
```

```python
df1 = pd.read_csv('data/tn.movie_budgets.csv', parse_dates = ['release_date'])
df1.head()
```

```python
df1.info()
```

```python
df1_clean = df1.copy()
```

```python
df1_clean['production_budget'] = df1_clean['production_budget'].str.replace(',', '').str.replace('$', '').astype(np.int64)
df1_clean['domestic_gross'] = df1_clean['domestic_gross'].str.replace(',', '').str.replace('$', '').astype(np.int64)
df1_clean['worldwide_gross'] = df1_clean['worldwide_gross'].str.replace(',', '').str.replace('$', '').astype(np.int64)
```

```python
df1_clean.drop(['id'], axis = 1, inplace = True)
```

```python
df1_clean.head()
```

This is now a clean data set.

```python
df2 = pd.read_csv('data/movie_profession.csv', index_col = 0)
df2.head()
```

```python
df2_clean = df2.copy()
```

```python
df2_clean.info()
```

```python
df2_clean['genres'] = df2_clean['genres'].fillna('Unknown')
df2_clean.head(10)
```
